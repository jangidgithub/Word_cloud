### Word cloud project

```python
%matplotlib inline
import wikipedia
import sys
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
from wordcloud import WordCloud,STOPWORDS
```

```python
# text = open("content.txt",'r').read()
```

```python
mask = np.array(Image.open("heart.png"))
# plt.imshow(mask)
```

```python
text = str(input("Enter the title-->"))
key = wikipedia.search(text)
page = wikipedia.page(key)
content = page.content


wc = WordCloud(stopwords=STOPWORDS,mask =mask,max_words=500,contour_color="black",background_color="white",
               contour_width=width=100,height = 100).generate(content)
plt.figure(figsize=(10,6))
plt.imshow(wc)
plt.axis("off")
plt.show()
```
